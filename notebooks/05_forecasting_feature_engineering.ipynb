```python
import pandas as pd
import numpy as np
```

```python
forecast_df = pd.read_csv(
    r"C:\Users\91991\Desktop\internships\AMDOX\data\processed\m5_forecasting_dataset.csv"
)
```

```python
forecast_df.shape
```

```python
forecast_df.shape
```

```python
forecast_df["date"] = pd.to_datetime(
    forecast_df["date"]
)
```

```python
forecast_df = forecast_df.sort_values(
    by=["id", "date"]
)
```

```python
forecast_df["lag_7"] = forecast_df.groupby(
    "id"
)["sales"].shift(7)
```

```python
forecast_df["lag_28"] = forecast_df.groupby(
    "id"
)["sales"].shift(28)
```

```python
forecast_df["rolling_mean_7"] = forecast_df.groupby(
    "id"
)["sales"].transform(
    lambda x: x.shift(1).rolling(7).mean()
)
```

```python
forecast_df["rolling_mean_28"] = forecast_df.groupby(
    "id"
)["sales"].transform(
    lambda x: x.shift(1).rolling(28).mean()
)
```

```python
forecast_df["day_of_week"] = forecast_df[
    "date"
].dt.dayofweek
```

```python
forecast_df["month"] = forecast_df[
    "date"
].dt.month
```

```python
forecast_df["week"] = forecast_df[
    "date"
].dt.isocalendar().week
```

```python
forecast_df["year"] = forecast_df[
    "date"
].dt.year
```

```python
forecast_df["is_event"] = forecast_df[
    "event_name_1"
].notnull().astype(int)
```

```python
forecast_df.head()
```

```python
forecast_df.head()
```

```python
forecast_df = forecast_df.dropna()
```
