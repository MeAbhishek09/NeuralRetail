```python
import pandas as pd
import numpy as np
```

```python
customer_df = pd.read_csv(
    r"C:\Users\91991\Desktop\internships\AMDOX\data\processed\clean_online_retail.csv"
)
```

```python
print(customer_df.shape)
customer_df.head()
```

```python
customer_df["InvoiceDate"] = pd.to_datetime(
    customer_df["InvoiceDate"]
)
```

```python
customer_df["InvoiceDate"].max()
```

```python
snapshot_date = customer_df[
    "InvoiceDate"
].max() + pd.Timedelta(days=1)

print(snapshot_date)
```

```python
rfm_df = customer_df.groupby(
    "Customer ID"
).agg({

    "InvoiceDate": lambda x: (
        snapshot_date - x.max()
    ).days,

    "Invoice": "nunique",

    "Revenue": "sum"

}).reset_index()
```

```python
rfm_df.columns = [
    "CustomerID",
    "Recency",
    "Frequency",
    "Monetary"
]
```
