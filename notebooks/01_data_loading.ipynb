```python
import pandas as pd


def clean_retail_data(file_path):

    df = pd.read_excel(file_path)

    df = df.drop_duplicates()

    df["InvoiceDate"] = pd.to_datetime(
        df["InvoiceDate"]
    )

    df = df.dropna(subset=["Customer ID"])

    df["Customer ID"] = df["Customer ID"].astype(int)

    df = df[df["Quantity"] > 0]

    df = df[df["Price"] > 0]

    df["Revenue"] = (
        df["Quantity"] * df["Price"]
    )

    return df
```
