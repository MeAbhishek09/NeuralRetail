```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    r"C:\Users\91991\Desktop\internships\AMDOX\data\processed\daily_business_sales.csv"
)

df["date"] = pd.to_datetime(df["date"])
```

```python
avg_daily_demand = df["sales"].mean()

std_daily_demand = df["sales"].std()

print("Average Demand:", avg_daily_demand)

print("Demand Std:", std_daily_demand)
```

```python
# Inventory assumptions

service_level = 0.95
z = 1.65
lead_time = 7


safety_stock = (
    z
    * std_daily_demand
    * np.sqrt(lead_time)
)

print("Safety Stock:", round(safety_stock))
```

```python
# Reorder point calculation
reorder_point = (
    avg_daily_demand * lead_time
) + safety_stock

print(
    "Reorder Point:",
    round(reorder_point)
)
```

```python
annual_demand = avg_daily_demand * 365

ordering_cost = 500

holding_cost = 50
```

```python
# Economic order quantity
eoq = np.sqrt(
    (2 * annual_demand * ordering_cost)
    / holding_cost
)

print(
    "EOQ:",
    round(eoq)
)
```
